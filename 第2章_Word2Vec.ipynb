```python
import gensim.downloader as api

corpus = api.load('text8')
```

```python
api.info("text8")
```

```python
from gensim.models import Word2Vec

model = Word2Vec(corpus)
print(model)
```

```python
similarity_words = model.wv.most_similar('japan')
similarity_words
```
